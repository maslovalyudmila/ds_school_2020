```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly
import plotly.graph_objs as go

init_notebook_mode(connected=True)
```

```python
# count of the number of charges and tenure by payment method
charges_df = df.groupby('PaymentMethod')[['MonthlyCharges']].sum().join(
    df.groupby('PaymentMethod')[['InternetService']].count()
)
charges_df.columns = ['tenure', 'MonthlyCharges']

# line for total charges
trace0 = go.Scatter(
    x=charges_df.index,
    y=charges_df.tenure,
    name='tenure'
)

# line for monthly charges
trace1 = go.Scatter(
    x=charges_df.index,
    y=charges_df.MonthlyCharges,
    name='MonthlyCharges'
)

# data array
data = [trace0, trace1]
layout = {'title': 'Statistics of charges'}

# create object figure and visualization
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
# count the number of loyal and disloyal churns of different payment methods
loyal_churns = df[df['Churn'] == 'No'].groupby(['PaymentMethod'])[['Churn']].count()
noloyal_churns = df[df['Churn'] == 'Yes'].groupby(['PaymentMethod'])[['Churn']].count()

loyal_churns.columns = ['Churn']
noloyal_churns.columns = ['Churn']

# create traces
trace0 = go.Bar(
    x=loyal_churns.index,
    y=loyal_churns.Churn,
    name='Loyal Churn'
)

trace1 = go.Bar(
    x=noloyal_churns.index,
    y=noloyal_churns.Churn,
    name='NoLoyal Churn'
)

# data array
data = [trace0, trace1]
layout = {'title': 'Churns', 'xaxis': {'title': 'PaymentMethod'}}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
# count the number of loyal and disloyal churns of different internet services
loyal_churns = df[df['Churn'] == 'No'].groupby(['InternetService'])[['Churn']].count()
notloyal_churns = df[df['Churn'] == 'Yes'].groupby(['InternetService'])[['Churn']].count()

loyal_churns.columns = ['Churn']
notloyal_churns.columns = ['Churn']

# create traces
trace0 = go.Bar(
    x=loyal_churns.index,
    y=loyal_churns.Churn,
    name='Loyal Churn'
)

trace1 = go.Bar(
    x=notloyal_churns.index,
    y=notloyal_churns.Churn,
    name='NoLoyal Churn'
)

# data array
data = [trace0, trace1]
layout = {'title': 'Churns', 'xaxis': {'title': 'InternetService'}}
fig = go.Figure(data=data, layout=layout)
iplot(fig, show_link=False)
```

```python
# create Box trace for every InternetService
data = []
for iService in df.InternetService.unique():
    data.append(
        go.Box(y=df[df.InternetService==iService].MonthlyCharges, name=iService)
    )

# data vizualization
iplot(data, show_link = False)
```
