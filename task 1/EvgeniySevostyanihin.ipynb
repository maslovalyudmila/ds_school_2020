## Feature engineering  
My idea is regroup wine for residual sugar and alcohol  

|name |alcohol |residual sugar |
|-----|:------:|:-------------:|
|dry  |8.5-15% |         < 0.3%|  
|medium dry |8.4-15% | 0.4-1.8%|
|semi-sweet|8.4-15% | 1.8-4.5%|
|sweet |8.4-15% | > 4.5%|
|fortified |17-21%|3-12%|
|fortified semi-sweet|14-16%|5-12%|
|fortified sweet|14-20%|> 15%|
|fortifed dessert|15-17%|16-20%|
|liquor|12-16%|21-30%|
|aromatised|16-18%|< 1.6%|
|sparkling|9-13%|< 1.5%|

```python
data['Wine Type'] = ''
```

```python
# i apologize for my code
data['Wine Type'].loc[(data['alcohol'] >= 8.4) & 
                  (data['alcohol'] < 15) &
                  (data['residual sugar'] < 0.3)] = 'dry'

data['Wine Type'].loc[(data['alcohol'] >= 8.4) & 
                  (data['alcohol'] < 15) &
                  (data['residual sugar'] > 0.4) &
                  (data['residual sugar'] <= 1.8)] = 'medium dry'

data['Wine Type'].loc[(data['alcohol'] >= 8.4) & 
                  (data['alcohol'] < 15) &
                  (data['residual sugar'] > 1.8) &
                  (data['residual sugar'] <= 4.5)] = 'semi-sweet'

data['Wine Type'].loc[(data['alcohol'] >= 8.4) & 
                  (data['alcohol'] < 15) &
                  (data['residual sugar'] > 4.5)] = 'sweet'

data['Wine Type'].loc[(data['alcohol'] > 17) & 
                  (data['alcohol'] < 21) &
                  (data['residual sugar'] > 3) &
                  (data['residual sugar'] < 12)] = 'fortified'

data['Wine Type'].loc[(data['alcohol'] > 14) & 
                  (data['alcohol'] < 16) &
                  (data['residual sugar'] > 5) &
                  (data['residual sugar'] < 12)] = 'fortified semi-sweet'

data['Wine Type'].loc[(data['alcohol'] > 14) & 
                  (data['alcohol'] < 20) &
                  (data['residual sugar'] > 15)] = 'fortified sweet'

data['Wine Type'].loc[(data['alcohol'] > 15) & 
                  (data['alcohol'] < 17) &
                  (data['residual sugar'] > 16) &
                  (data['residual sugar'] < 20)] = 'fortifed dessert'

data['Wine Type'].loc[(data['alcohol'] > 12) & 
                  (data['alcohol'] < 16) &
                  (data['residual sugar'] > 21) &
                  (data['residual sugar'] < 30)] = 'liquor'

data['Wine Type'].loc[(data['alcohol'] > 16) & 
                  (data['alcohol'] < 18) &
                  (data['residual sugar'] < 1.6)] = 'aromatised'

data['Wine Type'].loc[(data['alcohol'] > 9) & 
                  (data['alcohol'] < 13) &
                  (data['residual sugar'] < 1.5)] = 'sparkling'
```
